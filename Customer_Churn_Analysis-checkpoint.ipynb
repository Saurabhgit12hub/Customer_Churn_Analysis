```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

# replacing blanks with zero as tenure is 0 so no totalcharges recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

# converted 0 and 1 values of SeniorCitizen to yes/ no to make it easier for understanding

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn")
plt.show
```

#here i have done groupby Churn to count numbers

```python
gb = df.groupby("Churn").agg({'Churn':"count"})
gb
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.show()
```

# from given pie chart we can conclude that 26.54% of our Customer have churned out.
# now lets find reason behind this:


```python
plt.figure(figsize = (3,3))
ax = sns.countplot(x = "gender", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender")

plt.show()
```

```python
plt.figure(figsize = (3,3))
ax = sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender")

plt.show()
```

```python
crosstab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot stacked bar chart
ax = crosstab.plot(kind='bar', stacked=True, figsize=(4, 4), colormap='coolwarm')

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# Titles and labels
plt.title("Churn by Senior Citizen Status (%)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

plt.legend(title="Churn")
plt.show()
```

# comparatively a larger percentage of senior citizen category have churned out.

```python


# List of columns
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid (3 rows x 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Loop through each column and create a countplot
for i, col in enumerate(cols):
    ax = sns.countplot(x=col, data=df, hue='Churn', ax=axes[i], palette='Set2')
    ax.set_title(f'Countplot of {col}', fontsize=12)
    ax.set_xlabel('')  # Remove x-labels for cleaner look
    ax.set_ylabel('Count')

    # Add labels on bars
    for container in ax.containers:
        ax.bar_label(container, fontsize=9)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
